```python

df2 = pd.DataFrame ({
    'key': ['A','B','C','A','B','C'],
    'values': range(6)
},columns = ['key','values']
)
df2
```

```python
df2.groupby('key').mean()
```

Manipulacion de data frames


```python
df3 = pd.DataFrame({
    'key': ['A','B','C','A','B','C'],
    'value': range(6),
    'value2': np.random.randint(0,100,6)
})
```

```python
df3
```

```python
df3.groupby('key').aggregate(['min',np.median, np.mean,max])
```

```python
df3.groupby('key').aggregate({
    'value':'min',
    'value2':'max'
})
```

Filter

```python
df3.groupby('key').std()
```

```python
def filter_func(x):
  return x['value2'].std()>5
```

```python
df3.groupby('key').filter(filter_func)
```

**transform**

```python
df3.groupby('key').transform(lambda x: x -x.mean())
```

**Apply**

```python
def nor_by_col12(x):
      x['value'] = x['value']/ x['value2'].sum()
      return x
```

```python
df3.groupby('key').apply(nor_by_col12)
```
